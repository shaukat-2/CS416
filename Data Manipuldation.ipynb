```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head(10)
```

```python
df.country = df.country.str.replace(' ', '')
df.listed_in = df.listed_in.str.replace(' ', '')
```

```python
df['Country'] = df.country.str.split(',').tolist()
df['Genre'] = df.listed_in.str.split(',').tolist()
```

```python
new_df = df.explode('Country')
```

```python
new_df = new_df.dropna() 
```

```python
new_df[["show_id","Country"]].to_csv("countries.csv", index=False)
```

```python
new_df = df.explode('Genre')
new_df = new_df.dropna() 
```

```python
new_df[["show_id","Genre"]].to_csv("Genre.csv", index=False)
```
